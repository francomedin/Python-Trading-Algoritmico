````python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
````

````python
ds=pd.read_csv('/content/drive/MyDrive/Datos Colab/BTC_USD Bitfinex Historical Data.csv')
for i in ds:
  if i in ['Price','Open','High','Low']:
    ds[i]=ds[i].apply(lambda x: x.replace('.',''))
    ds[i] = ds[i].apply(lambda x: x.replace(',','.'))
    ds[i] = ds[i].astype(float)
  else:
    continue


ds
````

````python
#Convierto la columna Date
ds['Date']=pd.to_datetime(ds['Date'])
````

````python
ds=ds.set_index('Date')
ds=ds.sort_values('Date')
````

## Medias Moviles Simples

````python
from ta.trend import SMAIndicator
````

````python

sma50=SMAIndicator(ds['Price'],50,False)
sma100=SMAIndicator(ds['Price'],100,False)

ds['SMA-50']=sma50.sma_indicator()
ds['SMA-100']=sma100.sma_indicator()
ds
````

### Bandas Bollinger e Indicador ATR

````python
from ta.volatility import AverageTrueRange
````

````python
atr=AverageTrueRange(ds['High'],ds['Low'],ds['Price'],20,False)
````

````python
ds['ATR']=atr.average_true_range()
````

````python
ds
````

````python
fig = make_subplots(rows=2, cols=1, row_heights=[0.6,0.4], shared_xaxes=True, vertical_spacing=0.01)

fig.add_trace(go.Scatter(
    x=ds.index,
    y=ds['Price']),
    row=1,
    col=1
)

fig.add_trace(go.Scatter(
    x=ds.index,
    y=ds['ATR']),
    row=2,
    col=1
)

#Plot 1 texto
fig.update_yaxes(title_text='Precio BTC',row=1,col=1)
#Plot 2 texto
fig.update_yaxes(title_text='IndicadorATX',row=2,col=1)
fig.show()
````

````python
from ta.volatility import BollingerBands
````

````python
BandasBollinger=BollingerBands(ds['Price'],50,2,False)
````

````python
ds['BB B.Alta']=BandasBollinger.bollinger_hband()
ds['BB B.Baja']=BandasBollinger.bollinger_lband()
ds['BB B.Media']=BandasBollinger.bollinger_mavg()
ds['BB Altura']=BandasBollinger.bollinger_wband()

````

````python
ds.tail()
````

````python
fig2=make_subplots(rows=2,cols=1,row_heights=[0.6,0.4],shared_xaxes=True,vertical_spacing=0.01)

fig2.add_trace(go.Scatter(
    x=ds.index,
    y=ds['BB B.Alta'],
    line=dict(color='grey',width=1)
    ),
    row=1,
    col=1
)

fig2.add_trace(go.Scatter(
    x=ds.index,
    y=ds['BB B.Baja'],
    line=dict(color='blue',width=1),
    fill='tonexty'
    ),
    row=1,
    col=1
)

fig2.add_trace(go.Scatter(
    x=ds.index,
    y=ds['BB B.Media'],
    line=dict(color='red',width=1)
    ),
    row=1,
    col=1
)

fig2.add_trace(go.Scatter(
    x=ds.index,
    y=ds['Price']),
    row=1,
    col=1
)


fig2.add_trace(go.Scatter(
    x=ds.index,
    y=ds['ATR'],
     name='ATR'),
    row=2,
    col=1,
   
)

fig2.show()
````

## Indicador RSI

````python
from ta.momentum import rsi
````

````python
ds['RSI']=rsi(ds['Price'],20,False)
````

````python
ds
````

````python
fig3=make_subplots(rows=2,cols=1,row_heights=[0.7,0.3],shared_xaxes=True,vertical_spacing=0.1)

fig3.add_trace(go.Scatter(x=ds.index,y=ds['Price'],line=dict(color='grey',width=1)),row=1,col=1)
fig3.add_trace(go.Scatter(x=ds.index,y=ds['RSI']),row=2,col=1)

#Lineas horizontales del RSI
fig3.add_trace(go.Scatter(
    x=ds.index,
    y=[30]*len(ds['RSI']),
    name='RSI señal de compra',
    line=dict(color='black',width=1)),
    row=2,col=1
    )

fig3.add_trace(go.Scatter(
   x=ds.index,
    y=[70]*len(ds['RSI']),
    name='RSI señal de venta',
    line=dict(color='blue',width=1)),
    row=2,col=1
    )


fig3.update_yaxes(title='Precio btc',row=1,col=1)
fig3.update_yaxes(title='RSI',row=2,col=1)
fig3.update_layout(template='ggplot2')

fig3.show()
````

### Indicador ADX (Average Directional Index)

````python
 from ta.trend import ADXIndicator
````

````python
ADX=ADXIndicator(high=ds['High'],low=ds['Low'],close=ds['Price'],window=100,fillna= False)
````

````python
ds['ADX']=ADX.adx()
ds['MD+']=ADX.adx_pos()
ds['MD-']=ADX.adx_neg()
ds
````

````python
fig4=make_subplots(rows=2,cols=1,row_heights=[0.6,0.4],shared_xaxes=True, vertical_spacing=0.1)

fig4.add_trace(go.Scatter(
    x=ds.index,
    y=ds['Price'],
    name='BTC Price',
    line=dict(color='blue',width=1)),
    row=1,
    col=1
)

#Lineas direccionales + y -

fig4.add_trace(go.Scatter(
    x=ds.index,
    y=ds['MD+'],
    line=dict(color='purple',width=1.5),
    name='MD+'),
    row=2,
    col=1
)

fig4.add_trace(go.Scatter(
    x=ds.index,
    y=ds['MD-'],
    line=dict(color='black',width=1.5),
    name='MD-'),
    row=2,
    col=1
)


fig4.add_trace(go.Scatter(
    x=ds.index,
    y=ds['ADX'],
    name='Indicador ADX',
    line=dict(color='orange',width=1)),
    row=2,
    col=1
    )

fig4.update_yaxes(title_text='Precio del btc',row=1,col=1)
fig4.update_yaxes(title_text='Indicador ADX',row=2,col=1)

    
````
